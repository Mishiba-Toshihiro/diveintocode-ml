### ã€å•é¡Œ1ã€‘SimpleRNNã®ãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰ãƒ—ãƒ­ãƒ‘ã‚²ãƒ¼ã‚·ãƒ§ãƒ³å®Ÿè£…

ãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰ãƒ—ãƒ­ãƒ‘ã‚²ãƒ¼ã‚·ãƒ§ãƒ³ã®æ•°å¼ã¯ä»¥ä¸‹ã®ã‚ˆã†ã«ãªã‚Šã¾ã™ã€‚ndarrayã®shapeãŒã©ã†ãªã‚‹ã‹ã‚’ä½µè¨˜ã—ã¦ã„ã¾ã™ã€‚


ãƒãƒƒãƒã‚µã‚¤ã‚ºã‚’batch_sizeã€å…¥åŠ›ã®ç‰¹å¾´é‡æ•°ã‚’n_featuresã€RNNã®ãƒãƒ¼ãƒ‰æ•°ã‚’n_nodesã¨ã—ã¦è¡¨è¨˜ã—ã¾ã™ã€‚æ´»æ€§åŒ–é–¢æ•°ã¯tanhã¨ã—ã¦é€²ã‚ã¾ã™ãŒã€ã“ã‚Œã¾ã§ã®ãƒ‹ãƒ¥ãƒ¼ãƒ©ãƒ«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯åŒæ§˜ã«ReLUãªã©ã«ç½®ãæ›ãˆã‚‰ã‚Œã¾ã™ã€‚

$$

â„ğ‘¡âˆ’1htâˆ’1 : æ™‚åˆ»t-1ã®çŠ¶æ…‹ï¼ˆå‰ã®æ™‚åˆ»ã‹ã‚‰ä¼ã‚ã‚‹é †ä¼æ’­ï¼‰ (batch_size, n_nodes)


ğ‘Šâ„Wh : çŠ¶æ…‹ã«å¯¾ã™ã‚‹é‡ã¿ã€‚ (n_nodes, n_nodes)

ä¸Šè¨˜ã®å‡¦ç†ã‚’ç³»åˆ—æ•°n_sequenceså›ç¹°ã‚Šè¿”ã™ã“ã¨ã«ãªã‚Šã¾ã™ã€‚RNNå…¨ä½“ã¸ã®å…¥åŠ› ğ‘¥x ã¯(batch_size, n_sequences, n_features)ã®ã‚ˆã†ãªé…åˆ—ã§æ¸¡ã•ã‚Œã‚‹ã“ã¨ã«ãªã‚Šã€ãã“ã‹ã‚‰å„æ™‚åˆ»ã®é…åˆ—ã‚’å–ã‚Šå‡ºã—ã¦ã„ãã¾ã™ã€‚


åˆ†é¡å•é¡Œã§ã‚ã‚Œã°ã€ãã‚Œãã‚Œã®æ™‚åˆ»ã®hã«å¯¾ã—ã¦å…¨çµåˆå±¤ã¨ã‚½ãƒ•ãƒˆãƒãƒƒã‚¯ã‚¹é–¢æ•°ï¼ˆã¾ãŸã¯ã‚·ã‚°ãƒ¢ã‚¤ãƒ‰é–¢æ•°ï¼‰ã‚’ä½¿ç”¨ã—ã¾ã™ã€‚ã‚¿ã‚¹ã‚¯ã«ã‚ˆã£ã¦ã¯æœ€å¾Œã®æ™‚åˆ»ã®hã ã‘ã‚’ä½¿ç”¨ã™ã‚‹ã“ã¨ã‚‚ã‚ã‚Šã¾ã™ã€‚

```python
import numpy as np
```

```python
class SimpleRNN:
    """
    RNNä¸­é–“å±¤
    Parameters
    ----------
    n_features : int
      RNNã®å…¥åŠ›ã®æ¬¡å…ƒ
    n_nodes : int
      RNNã®ãƒãƒ¼ãƒ‰æ•°
    initializer : åˆæœŸåŒ–æ–¹æ³•ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹
    optimizer : æœ€é©åŒ–æ‰‹æ³•ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹

    Attribute
    --------
    self.W_x : ndarray(n_features, n_nodes)
      å…¥åŠ›ã«å¯¾å¿œã™ã‚‹é‡ã¿
    self.W_h : ndarray(n_nodes, n_nodes)
      çŠ¶æ…‹ã«å¯¾å¿œã™ã‚‹é‡ã¿
    self.B : ndarray(n_node,)
      ãƒã‚¤ã‚¢ã‚¹
    self.X_ : ndarray(batch_size, n_features)
      ä¿å­˜ã•ã‚ŒãŸå…¥åŠ›
    """
    def __init__(self, n_features, n_nodes, initializer=None, lr=0.01):
        # åˆæœŸåŒ–
        # initializerã®ãƒ¡ã‚½ãƒƒãƒ‰ã‚’ä½¿ã„ã€self.Wx, self.Wt, self.Bã‚’åˆæœŸåŒ–ã™ã‚‹
        if initializer is not None:
            self.W_x = initializer.W(n_features, n_nodes)
            self.W_h = initializer.W(n_nodes,    n_nodes)
            self.B   = initializer.B(n_nodes)
        self.lr = lr
        
    def forward(self, h_t, X):
        """
        ãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰
        Parameters
        ----------
        h_t : ndarray(batch_size, n_nodes)
            çŠ¶æ…‹
        X : ndarray(batch_size, n_features)
            å…¥åŠ›
        Returns
        ----------
        A : ndarray(batch_size, n_nodes)
            å‡ºåŠ›
        """
        self.X_ = X

        A = X@self.W_x + h_t@self.W_h + self.B

        return A
```

### ã€å•é¡Œ2ã€‘å°ã•ãªé…åˆ—ã§ã®ãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰ãƒ—ãƒ­ãƒ‘ã‚²ãƒ¼ã‚·ãƒ§ãƒ³ã®å®Ÿé¨“
å°ã•ãªé…åˆ—ã§ãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰ãƒ—ãƒ­ãƒ‘ã‚²ãƒ¼ã‚·ãƒ§ãƒ³ã‚’è€ƒãˆã¦ã¿ã¾ã™ã€‚


å…¥åŠ›xã€åˆæœŸçŠ¶æ…‹hã€é‡ã¿w_xã¨w_hã€ãƒã‚¤ã‚¢ã‚¹bã‚’æ¬¡ã®ã‚ˆã†ã«ã—ã¾ã™ã€‚


ã“ã“ã§é…åˆ—xã®è»¸ã¯ãƒãƒƒãƒã‚µã‚¤ã‚ºã€ç³»åˆ—æ•°ã€ç‰¹å¾´é‡æ•°ã®é †ç•ªã§ã™ã€‚



```python
x = np.array([[[1, 2], [2, 3], [3, 4]]])/100 # (batch_size, n_sequences, n_features)
w_x = np.array([[1, 3, 5, 7], [3, 5, 7, 8]])/100 # (n_features, n_nodes)
w_h = np.array([[1, 3, 5, 7], [2, 4, 6, 8], [3, 5, 7, 8], [4, 6, 8, 10]])/100 # (n_nodes, n_nodes)
batch_size = x.shape[0] # 1
n_sequences = x.shape[1] # 3
n_features = x.shape[2] # 2
n_nodes = w_x.shape[1] # 4
h = np.zeros((batch_size, n_nodes)) # (batch_size, n_nodes)
b = np.array([1, 1, 1, 1]) # (n_nodes,)
```

```python
class Tanh:
    """
    tanhé–¢æ•°

    Attribute
    ----------
    self.A
        æ´»æ€§åŒ–é–¢æ•°ã®å…¥åŠ›
    """
    def forward(self, A):
        self.A = A
        Z = (np.exp(A)-np.exp(-A)) / (np.exp(A)+np.exp(-A))
        
        return Z
```

```python
sampleRNN = []

# n_sequencesæ–¹å‘ã«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ä½œæˆ
for i in range(n_sequences):
    sampleRNN.append([SimpleRNN(n_features=n_features, n_nodes=n_nodes), Tanh()])
    # é‡ã¿ã¨ãƒã‚¤ã‚¢ã‚¹ã®è¨­å®š
    sampleRNN[i][0].W_x = w_x
    sampleRNN[i][0].W_h = w_h
    sampleRNN[i][0].B   = b

# é †ä¼æ¬
h_tmp = h
for j in range(n_sequences):
    # RNNå±¤
    h_tmp = sampleRNN[j][0].forward(h_tmp, x[:, j, :])
    # æ´»æ€§åŒ–é–¢æ•°
    h_tmp = sampleRNN[j][1].forward(h_tmp)
h_last = h_tmp

"""
æœ€çµ‚çš„ãªå‡ºåŠ›ã®ç¢ºèª
output:
h = np.array([[0.79494228, 0.81839002, 0.83939649, 0.85584174]]) # (batch_size, n_nodes)
[[0.79494228, 0.81839002, 0.83939649, 0.85584174]] (1, 4)
"""
print(h_last, h_last.shape)
```
