欠損値は大きく分けて3種類のパターンがあります。それぞれ欠損値への対応に違いがでてきます。


**Missing Completely At Random(MCAR)**

完全にランダムに欠損。

欠損に理由や意味はない場合。（例）記入漏れ。

**Missing At Random(MAR)**

観測データによる欠損。

他の特徴量の値を理由に欠損している場合。（例）若年層は答えていない項目。

**Missing Not At Random(MNAR)**

欠損データによる欠損。

欠損値の本来の値を理由に欠損している場合。（例）温度が上がりすぎてセンサーが測れない。

### 【問題7】特徴量間の関係の確認

　特徴量間の関係を見ていきます。次の4つを行ってください。

13 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
# x="sepal length (cm)", y="sepal width (cm)"の散布図を表示
sns.lmplot(x="sepal length (cm)", y="sepal width (cm)",hue="Species",data=df)
plt.title("sepal length (cm) - sepal width (cm)")
```

```python
# x="sepal length (cm)", y="petal length (cm)の散布図を表示
sns.lmplot(x="sepal length (cm)", y="petal length (cm)", hue="Species", data=df)
plt.title("sepal length (cm) - petal length (cm)")
```

```python
# x="sepal length (cm)", y="petal width (cm)の散布図を表示
sns.lmplot(x="sepal length (cm)", y="petal width (cm)", hue="Species", data=df)
plt.title("sepal length (cm) - petal width (cm)")
```

```python
# x="sepal width (cm)", y="petal length (cm)の散布図を表示
sns.lmplot(x="sepal width (cm)", y="petal length (cm)", hue="Species", data=df)
plt.title("sepal width (cm) - petal length (cm)")
```

```python
# x="sepal width (cm)", y="petal width (cm)の散布図を表示
sns.lmplot(x="sepal width (cm)", y="petal width (cm)", hue="Species", data=df)
plt.title("sepal width (cm) - petal width (cm)")
```

```python
# x="petal length (cm)", y="petal width (cm)の散布図を表示
sns.lmplot(x="petal length (cm)", y="petal width (cm)", hue="Species", data=df)
plt.title("petal length (cm) - petal width (cm)")
```

```python
#参考
# ちなみに上記のコードに　col="Species"　を加えると違うグラフに表示できる
sns.lmplot(x="petal length (cm)", y="petal width (cm)", hue="Species",col="Species", data=df,aspect=1)
plt.title("petal length (cm) - petal width (cm)")
```

14 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

```python
sns.pairplot(df,hue='Species')
```

15 ４つの特徴量に関する相関係数行列を作る

```python
corr_matrix = df.corr()
print(corr_matrix)
```

16 相関係数行列のヒートマップを作る

```python
sns.heatmap(corr_matrix,square=True,vmax=1.0,vmin=-1.0)
```

```python
# 特徴量とラベル間の相関係数のみ抽出して見る
corr_matrix["Species"]
# petal length(cm)とpetal width(cm)がそれぞれSpeciesとの相関係数0.9以上のため強い相関がありそうである。
```
